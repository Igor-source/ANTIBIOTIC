```python
def convert_to_list(x):
    if isinstance(x, str):
        x = x.strip().strip("'\"")  # Удаляем лишние пробелы и внешние кавычки
        try:
            x = ast.literal_eval(x)
        except Exception as e:
            print(f"Ошибка при преобразовании в список на значении {x}: {e}")
            x = np.nan
    return x
```

```python
def remove_first_value(x):
    if isinstance(x, (list, ListElement)):
        if len(x) > 1:
            return x[1]  # Возвращаем второй элемент списка
        elif len(x) == 1:
            return x[0]  # Если только один элемент, возвращаем его
        else:
            return np.nan
    else:
        return x  # Возвращаем значение как есть, если оно не список
```

```python

def get_tax_id(bacterial_name):
    """Fetches the Taxonomy ID for a given bacterial name."""
    try:
        search_handle = Entrez.esearch(db="taxonomy", term=bacterial_name)
        search_results = Entrez.read(search_handle)
        search_handle.close()

        # Get the first Taxonomy ID from the search results
        if search_results["IdList"]:
            return search_results["IdList"][0]
        else:
            print(f"No Taxonomy ID found for {bacterial_name}")
            return None
    except Exception as e:
        print(f"Error searching for {bacterial_name}: {e}")
        return None

def get_superkingdom_from_name(bacterial_name):
    """Fetches the superkingdom for a given bacterial name."""
    tax_id = get_tax_id(bacterial_name)
    if tax_id:
        try:
            handle = Entrez.efetch(db="taxonomy", id=tax_id, retmode="xml")
            record = Entrez.read(handle)
            handle.close()

            if record:
                for lineage in record[0]['LineageEx']:
                    if lineage['Rank'] == 'superkingdom':
                        return lineage['ScientificName']
            return None
        except Exception as e:
            print(f"Error fetching superkingdom for Taxonomy ID {tax_id}: {e}")
            return None
    return None

# Create a new column 'Superkingdom' in dfr
dfr['Superkingdom'] = None

# Iterate through the rows and fetch the superkingdom for each bacterial name in the 'name' column
for index, row in dfr.iterrows():
    bacterial_name = row['name']

    # Fetch the superkingdom and add delay
    superkingdom = get_superkingdom_from_name(bacterial_name)
    dfr.at[index, 'Superkingdom'] = superkingdom
    time.sleep(0.4)  # Delay to comply with NCBI rate limits

dfr
```

```python
# Создаем пустой список для хранения уровня таксономической классификации "Kingdom"
kingdom = []

# Создаем пустые DataFrame для бактерий и грибов
bacteria_df = pd.DataFrame()
fungi_df = pd.DataFrame()

# Итерация по DataFrame
for i in range(len(dfr)):
    id_value = dfr['IdList'].iloc[i]

    # Проверяем, является ли ID допустимым
    if pd.isna(id_value) or not isinstance(id_value, (str, int)):
        print(f"Некорректный ID на индексе {i}: {id_value}")
        kingdom.append(np.nan)
        continue

    id_value = str(id_value)  # Преобразуем ID в строку

    try:
        handle = Entrez.efetch(db="Taxonomy", id=id_value, retmode="xml")
        records = Entrez.read(handle)
        handle.close()

        # Получаем таксономическую линию
        tax = records[0]["Lineage"].split("; ")
        time.sleep(0.2)

        # Проверяем, являются ли это грибами или бактериями
        if any("Fungi" in tax_elem for tax_elem in tax):
            # Добавляем уровень "Kingdom" в список
            kingdom.append(tax[3] if len(tax) > 3 else np.nan)
            # Добавляем данные в DataFrame грибов
            fungi_df = pd.concat([fungi_df, dfr.iloc[[i]]], ignore_index=True)
        elif any("Bacteria" in tax_elem for tax_elem in tax):
            # Добавляем уровень "Kingdom" в список
            kingdom.append(tax[1] if len(tax) > 1 else np.nan)
            # Добавляем данные в DataFrame бактерий
            bacteria_df = pd.concat([bacteria_df, dfr.iloc[[i]]], ignore_index=True)
        else:
            kingdom.append(np.nan)
    except Exception as e:
        print(f"Ошибка при обработке ID {id_value} на индексе {i}: {e}")
        kingdom.append(np.nan)

# Создаем новый столбец в DataFrame для уровня классификации "Kingdom"
dfr["Kingdom"] = kingdom

# Сохраняем модифицированный DataFrame в новый файл
dfr.to_csv("taxonomy_kingdom.csv", index=False)

# Выводим столбец Kingdom
print(dfr["Kingdom"])
```
