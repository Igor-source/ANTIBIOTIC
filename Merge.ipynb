```python
# Список единиц измерения, не относящихся к концентрации
not_concentration_units = [
    'mm', 'hr', 'um', '%', '/s', 'ug/sq.cm', 'ucm/s', 'min',
    'mm/mg', 'mg', 'cm', 'umol', 'ug', 'nmol'
]

# Фильтруем DataFrame, оставляя только строки, где значение в столбце 'units' не содержит этих единиц
drop_columns = drop_columns[~drop_columns['units'].isin(not_concentration_units)]
```

```python
drop_columns.reset_index(drop=True, inplace=True)
drop_columns
```

```python
drop_columns = drop_columns.dropna(subset=["units"])
```

```python
import re

def convert_to_ug_ml(row):
    unit = row['units']
    value = row['value']
    mol_wt = row.get('MolWt', None)  # Молекулярная масса в г/моль, если она нужна для конверсии

    if pd.isna(unit):  # Проверка на NaN в единицах измерения
        return None

    if unit == 'ug ml-1':
        return value
    elif unit == 'mg/ml':
        return value * 1000
    elif unit == 'g ml-1':
        return value * 1_000_000
    elif unit in ['mg/L', 'mg l-1']:
        return value
    elif unit == 'mg/dL':
        return value * 10
    elif unit == 'uM' and mol_wt is not None:
        return value * mol_wt
    elif unit == 'mM' and mol_wt is not None:
        return value * mol_wt * 1000
    elif unit == 'mmol/L' and mol_wt is not None:
        return value * mol_wt
    elif unit == 'nM' and mol_wt is not None:
        return value * mol_wt * 0.001
    elif unit == 'pg/ml':
        return value * 0.001
    elif unit == 'ng/ml' or unit == 'ng ml-1':
        return value * 0.001
    elif unit == 'g/dL':
        return value * 10_000
    elif unit == 'uM/L' or unit == 'uM l-1':
        return value * mol_wt if mol_wt is not None else None
    elif unit == 'umol/L' or unit == 'umol/ml' or unit == "10'-6 mol/L":
        return value * mol_wt if mol_wt is not None else None
    elif unit == 'ppm':
        return value * 1  # 1 ppm = 1 mg/L
    elif unit == 'M' and mol_wt is not None:
        return value * mol_wt * 1_000_000
    # Обработка единиц вида "10'-3 uM/ml"
    elif re.match(r"10'-\d+", unit):  # Проверка на соответствие формату "10'-N"
        try:
            multiplier = 10 ** int(re.search(r"10'-([\d-]+)", unit).group(1))
            if 'uM' in unit and mol_wt is not None:
                return value * mol_wt * multiplier
            elif 'mM' in unit and mol_wt is not None:
                return value * mol_wt * multiplier * 1000
            elif 'umol' in unit and mol_wt is not None:
                return value * mol_wt * multiplier
        except (ValueError, AttributeError):
            return None
    else:
        return None  # Если нет определённого коэффициента перевода

# Применение функции к DataFrame
drop_columns['value_in_ug_ml'] = drop_columns.apply(convert_to_ug_ml, axis=1)
```
